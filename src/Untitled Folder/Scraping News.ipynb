```python

def read_file():
    df = pd.read_csv('../../Data/Raw_Newspapper.csv', names=['Name', 'City','URL'])
    return df
```

```python

def make_soup(url):
    response = rq.get(url)
    if response.status_code == 200:
        html_text = response.text
        soup = BeautifulSoup(html_text, "html.parser")
    else:
        print("NA")
    return soup
        
```

```python
df = read_file()
```

```python
newspaper_responses = []
example = make_soup(df['URL'][1])
example
```

```python
links =[]
for _ in df['URL']:
    links.append(_)
links
    
```
